```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('dataset.csv')
df
```

```python
notas_por_materia = [df[df['materia'] == materia]['nota'].tolist() for materia in df['materia'].unique()]

plt.figure(figsize=(10, 6))
plt.boxplot(notas_por_materia, labels=df['materia'].unique())

plt.title('Distribución de Notas')
plt.xlabel('Materias')
plt.ylabel('Nota')
plt.show()
```

```python
# Contar la cantidad de aprobados y no aprobados
aprobados = df[df['aprobado'] == 'Sí'].shape[0]
no_aprobados = df[df['aprobado'] == 'No'].shape[0]

plt.figure(figsize=(6, 6))
plt.pie([aprobados, no_aprobados], labels=['Aprobados', 'No Aprobados'], autopct='%1.1f%%', startangle=90, colors=['dodgerblue', 'orangered'])

plt.title('Distribucion de Aprobados')

plt.show()
```
